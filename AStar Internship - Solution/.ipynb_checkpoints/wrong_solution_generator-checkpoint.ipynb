## Loading the error list

```python
#Load errors that are specific to the topics
def load_specific_error_list(mode = "Specific"):
    if mode == "Specific":
        df = pd.read_excel('./dataset/topic_specific_errors_only.xlsx')
    else:
        df = pd.read_excel('./dataset/topic_specific_errors.xlsx', sheet_name = mode)
        
    error_types = {}
    
    for _, row in df.iterrows():
        key1 = row['topic']
        key2 = row['error']
        value = row['explanation']
        
        if key1 not in error_types:
            error_types[key1] = {}
            error_types[key1]['type_of_error'] = []
            error_types[key1]['error_explanation'] = {}
        
        error_types[key1]['type_of_error'].append(key2)
        error_types[key1]['error_explanation'][key2] = value
    return error_types
error_types = load_specific_error_list()
print(error_types['P5N1m'])
```

```python
# If we use general errors only
def load_general_error_list():
    df = pd.read_excel('./dataset/new_general_errors.xlsx')
    
    error_types = {}
    error_types['type_of_error'] = []
    error_types['error_explanation'] = {}
    
    for _, row in df.iterrows():
        error_types['type_of_error'].append(row['Error name'])
        error_types['error_explanation'][row['Error name']] = row['Description']
    return error_types

error_types = load_general_error_list()
print(error_types.keys())
```

```python
#Append the specific errors onto the general errors
def combine_error_list(question_idx):
    general_error_types = load_general_error_list()
    specific_error_types = load_specific_error_list()
    combined_list = general_error_types.get('type_of_error', None).copy()
    combined_explanation = general_error_types.get('error_explanation',None).copy()
    #Append errors from topic-specific error list
    if specific_error_types.get(question_idx, None) is not None:
        combined_list.extend(specific_error_types[question_idx]['type_of_error'])
        for key in specific_error_types[question_idx]['error_explanation'].keys():
            combined_explanation[key] = specific_error_types[question_idx]['error_explanation'][key]
    combined_dct = {'type_of_error' : combined_list,
                    'error_explanation' : combined_explanation}
    return combined_dct
```

```python
combined_error_types = {}
for question_idx in dataset.keys():
    combined_error_types[question_idx] = combine_error_list(question_idx)

combined_error_types_fpath = './solution/wrong solution/step 2/combined_error_types.json'
with open(combined_error_types_fpath,'w') as file:
    json.dump(combined_error_types, file, indent = 4)
```

## Generating worked example

```python
"""
This code block is used to generate wrong solutions to questions based on the list of error types used earlier

Variables in this code block:
Input:
dataset (dct) : { (question ID) : {'ID' : question ID,
                                   'Grade' : the grade the question is in,
                                   'KC' : knowledge component of the question,
                                   'question' : question content
                                   } : Question dataset
solution (dct): { (question ID) : {'question' : question content,
                                   'kc' : knowledge component of this question,
                                   'solution' : {'step (number)' : step content}
                                  }
                } : Correct solution to the question
general_error_types (dct): {'type_of_error' : full list of error types,
                    'error_explanations' (dct) : {error_type : error_explanation},
                    'root_cause' (dct) : {error_type : root_cause},
                    'how_to_avoid' (dct) : {error_type : how_to_avoid}
                    } : Error types that applies to almost all topics
specific_error_types (dct) : {(topicID) : {dict with keys ['type_of_error','error_explanation','root_cause','how_to_avoid']}}
                           : Error types that is specific to the particular topic
models : List of large language models to be used
system_prompt_fpath : System prompt text file path
user_prompt_fpath : User prompt text file path
generated_solution_fpath : File path to save generated solutions as a json file in
random_sample_question_mode : Turn on/off random sample mode. If random sample mode is turned on (True), then we choose randomly topic_sample_size topics and
Else (False), we sample every .
random_sample_ertype_mode : Turn on/off random sample mode. If random sample mode is turned on (True), then we choose randomly ertype_sample_size topics.
Else (False), we sample every types of error.
topic_sample_size : Number of topics to be randomly sampled. If larger than total number of topics, reduced to total number of topics
ertype_sample_size : Number of error types for each topic to be randomly sampled. If larger than total number of errors, reduced to total number of errors
step_to_be_injected : The step in the original solution that needs to be injected. Set to 999 if random placement of error is wanted.

Output:
wrong_solution_dct (dct): Dictionary to store generated wrong solutions
wrong_solution_text (dct): Dictionary to store original LLM's response (for debugging)
errors (dct): Errors happening during response generation and dictionary type conversion
A json file saved  in generated_solution_fpath
"""
#Inputs
#dataset loaded earlier in initialization (before step 1)
#models loaded earlier in initialization (before step 1)
#solutions loaded earlier in initialization (before step 1)
general_error_types = load_general_error_list()
#specific_error_types = load_specific_error_list()
random_sample_question_mode = False
random_sample_ertype_mode = False
topic_sample_size = 3
ertype_sample_size = 10
step_to_be_injected = 99

#Random sample the topics, using topic ID
if random_sample_question_mode and topic_sample_size < len(dataset.keys()):
    question_list = random.sample(list(dataset.keys()),topic_sample_size)
else:
    question_list = list(dataset.keys())
question_len = len(question_list)
remaining_qn_list = list(set(dataset.keys()) - set(question_list))

all_error_lists = {}

#Random samples the types of error for each question
for question_idx in question_list:
    if random_sample_ertype_mode and ertype_sample_size < len(combined_error_types[question_idx]['type_of_error']):
        error_list = random.sample(combined_error_types[question_idx]['type_of_error'],ertype_sample_size)
    else:
        error_list = combined_error_types[question_idx]['type_of_error']

    all_error_lists[question_idx] = error_list

for model in models[:1]:
    print("Currently working with model: ", model)
    #Replace / as _ so it wouldn't interfere with path directory
    model_path = model.replace("/","_")
    system_prompt_fpath = f"./prompt/wrong solution/simple_prompts/{model_path}_system_prompt.txt" 
    user_prompt_fpath = f"./prompt/wrong solution/simple_prompts/{model_path}_user_prompt.txt"
    generated_solution_fpath = f"./solution/wrong solution/dataset_{file_id}/simple_prompts/{model_path}_wrong_solution.json"
    
    wrong_solution_dct = {}
    wrong_solution_text = {}
    runtime_errors = {}

    question_idx_num = 0
    for question_idx in question_list:
        question_idx_num += 1

        info = {
            "question" : dataset[question_idx]["response_dict"]["word_problem"],
            "answer" : dataset[question_idx]["response_dict"]["solution"],
            "primary_kc" : dataset[question_idx]["Primary_kc"],
            "secondary_kc" : dataset[question_idx]["Secondary_kc"],
            "grade" : dataset[question_idx]["Grade"]
        }
        
        error_idx = 0
        for type_of_error in all_error_lists[question_idx]:
            error_idx += 1

            worked_example_ID = question_idx + '_' + type_of_error
            
            info['error_explanation'] = combined_error_types[question_idx]['error_explanation'][type_of_error]
            #Load prompt for solution generation
            try:
                with open(system_prompt_fpath,"r") as file:
                    system_prompt = file.read()
            except Exception as e:
                system_prompt = ""
            try:
                with open(user_prompt_fpath,"r") as file:
                    user_prompt = file.read()
            except Exception as e:
                user_prompt = ""

            this_system_prompt = system_prompt.format()
            this_user_prompt = user_prompt.format(type_of_error = info["error_explanation"], primary_kc = info["primary_kc"], 
                                                  secondary_kc = info["secondary_kc"], question = info["question"], solution = info["answer"], 
                                                  grade = info["grade"])
            
            #Different models are called by different API
            if model in models_azure:
                wrapper = Azure()
            elif model in models_aws:
                wrapper = Aws()
            else:
                wrapper = OpenRouter()
    
            if model == "openai/gpt-4.1":
                class MWE(BaseModel):
                    worked_example: str
                    error_explanation: str
                message = wrapper.get_GPT_struct_response(this_system_prompt, this_user_prompt, MWE)
                wrong_solution_dct[worked_example_ID] = message
                
                #Progress tracker
                print(f"\rQuestion {question_idx} ({question_idx_num}/{question_len}), error type {error_idx}/{len(all_error_lists[question_idx])} generated", end = '', flush = True)
                continue
            else:
                message = wrapper.get_LLM_response(models_platform_translation[model], this_system_prompt, this_user_prompt)
        
            response_text = message.strip()
            if model == "deepseek/deepseek-r1":
                _,__,response_text = response_text.partition("</think>")
                response_text = response_text.strip()    
            response_text_extracted = None

            #Extract and convert to dictionary
            match = re.findall(r'\{.*\}', response_text, re.DOTALL)
            if match:
                response_text_extracted = match[0]
            else:
                print("NO JSON FORMAT OUTPUT FOUND! ERROR!")
                print(response_text)   
            try:
                response_dct = ast.literal_eval(response_text_extracted)
                wrong_solution_dct[worked_example_ID] = response_dct
            except Exception as e:
                runtime_errors[question_idx][type_of_error] = response_text
                print("Found error at", question_idx, type_of_error, "error: ", e)
                print(response_text)
            #Save original response text for debugging
            wrong_solution_text[question_idx][type_of_error] = response_text
            

            #Progress tracker
            print(f"\rQuestion {question_idx} ({question_idx_num}/{question_len}), error type {error_idx}/{len(all_error_lists[question_idx])} generated", end = '', flush = True)


    #Save to file
    os.makedirs(os.path.dirname(generated_solution_fpath), exist_ok=True)
    with open(generated_solution_fpath, 'w') as file:
        json.dump(wrong_solution_dct, file, indent=4)
```

```python
def fix_latex_error_str(s):
    s = s.replace("\x0crac","\\frac")
    s = s.replace('\n', ' \n ')   # Replace \n with LaTeX newline command
    s = s.replace('$', '\\$')
    s = s.replace('%', '\\%')
    s = s.replace('\\\\', '\\')   # Normalize double backslashes
    return s

def fix_latex_error_dct(dct):
    for key in dct.keys():
        if isinstance(dct[key],dict):
            dct[key] = fix_latex_error_dct(dct[key])
        elif isinstance(dct[key], str):
            dct[key] = fix_latex_error_str(dct[key])
        else:
            print(f"Error at {dct[key]}: Function not supported for type: {type(dct[key])}")
    return dct

print(fix_latex_error_dct(wrong_solution_dct))
```

## Step 4: Generating multiple error type solutions

```python
def load_extract_wrong_solution(generated_solution_path = None, dct = None):
    if dct is None:
        if generated_solution_path is None:
            print("Error: No data (file path or dictionary) given. Returning...")
            return
        with open(generated_solution_path, 'r') as file:
            dct = json.load(file)
    for kc_id in dct.keys():
        for error_type in dct[kc_id].keys():
            # Parse step numbers and sort keys accordingly
            sorted_items = sorted(dct[kc_id][error_type]["wrong_steps"].items(), key=lambda x: int(x[0].split()[1]))
            
            seen_values = set()
            result = {}
            
            for key, value in sorted_items:
                if value not in seen_values:
                    seen_values.add(value)
                    result[key] = value
            
            dct[kc_id][error_type]["wrong_steps"] = result
    return dct

load_extract_wrong_solution()
```

```python
"""
Variables in this code block:
Input:
dataset : Question dataset
solution : Correct solutions to the questions
wrong_solution : Previously wrong solution
general_error_types (dct): {'type_of_error' : Full list of error types,
                    'error_explanations' (dct) : {error_type : error_explanation},
                    'root_cause' (dct) : {error_type : root_cause},
                    'how_to_avoid' (dct) : {error_type : how_to_avoid}
                    } : Error types that applies to almost all topics
specific_error_types (dct) : {(topicID) : {dict with keys ['type_of_error','error_explanation','root_cause','how_to_avoid']}}
                           : Error types that is specific to the particular topic
models : List of large language models to be used
system_prompt_fpath : System prompt text file path
user_prompt_fpath : User prompt text file path
generated_solution_fpath : File path to save generated solution
random_sample_mode : Turn on/off random sample mode. If random sample mode is turned on (True), then we choose randomly topic_sample_size topics and
ertype_sample_size error types. Else (False), we sample every topics and every types of error
topic_sample_size : Number of topics to be randomly sampled. If larger than total number of topics, reduced to total number of topics
ertype_sample_size : Number of error types to be randomly sampled. If larger than total number of errors, reduced to total number of errors

Output:
wrong_solution_dct (dct): Dictionary to store generated wrong solutions
wrong_solution_text (dct): Dictionary to store original LLM's response (for debugging reasons)
errors (dct): Errors happening during response generation and dictionary type conversion
A json file saved  in generated_solution_fpath
"""
#Inputs
#dataset loaded earlier in initialization (before step 1)
#models loaded earlier in initialization (before step 1)
#solution loaded earlier in initialization (before step 1)
#wrong_solution loaded in the previous cell
wrong_solution = sampled_wrong_solution
general_error_types = load_general_error_list()
specific_error_types = load_specific_error_list()
random_sample_mode = True
topic_sample_size = 7
ertype_sample_size = 1

for model in models[0:1]:
    print("Currently working with model: ", model)
    #Replace / as _ so it wouldn't interfere with path directory
    model_path = model.replace("/","_")
    system_prompt_fpath = f"./prompt/wrong solution/step 4. injecting extra error types/{model_path}_system_prompt.txt" 
    user_prompt_fpath = f"./prompt/wrong solution/step 4. injecting extra error types/{model_path}_user_prompt.txt" 
    generated_solution_fpath = f'./solution/wrong solution/step 4/{model_path}_wrong_solution.json'
    
    wrong_solution_dct = {}
    wrong_solution_text = {}
    runtime_errors = {}

    #Random sample the topics, using topic ID
    if random_sample_mode and topic_sample_size < len(wrong_solution.keys()):
        question_list = random.sample(list(dataset.keys()),topic_sample_size)
    else:
        question_list = list(wrong_solution.keys())
    question_len = len(question_list)

    question_idx_num = 0
    for question_idx in question_list:
        question_idx_num += 1
        wrong_solution_text[question_idx] = {}
        wrong_solution_dct[question_idx] = {}
        runtime_errors[question_idx] = {}

        info = {
            "question" : dataset[question_idx]["Question"],
            "answer" : solution[question_idx]["solution"],
            "wrong_answer" : wrong_solution[question_idx],
            #Try to get 'type_of_error' from general error list, then append the topic-specific list later
            #This is to "broad_cast" the general error types to all topics
            "error_list" : combine_error_list(question_idx)['type_of_error'],
            "error_explanation" : combine_error_list(question_idx)['error_explanation']
        }

        latest_wrong_step = int(list(wrong_solution[question_idx]["all_wrong_steps"].keys())[-1].split()[1])
        
        #Random samples the types of error
        if random_sample_mode and ertype_sample_size < len(info['error_list']):
            error_list = random.sample(info['error_list'],ertype_sample_size)
        else:
            error_list = info['error_list']
        #Latest wrong step is too close to final step
        if latest_wrong_step >= len(wrong_solution[question_idx]["wrong_solution"]):
            print("The latest wrong step is too close to the end already!")
            continue
        
        error_idx = 0
        for type_of_error in error_list:
            error_idx += 1
            
            #Skip the error type if the wrong solution contains that already
            if type_of_error in info['wrong_answer'].get('all_wrong_steps',{}).values():
                print("This type of error already exist in the solution")
                wrong_solution_dct[question_idx] = wrong_solution[question_idx].copy()
                wrong_solution_dct[question_idx]["all_wrong_steps"] = wrong_solution[question_idx]["wrong_steps"].copy()
                wrong_solution_dct[question_idx]["all_explanations"] = wrong_solution[question_idx]["error_explanation"].copy()
                continue
            
            #Load prompt for solution generation
            try:
                with open(system_prompt_fpath,"r") as file:
                    system_prompt = file.read()
            except Exception as e:
                system_prompt = ""
            try:
                with open(user_prompt_fpath,"r") as file:
                    user_prompt = file.read()
            except Exception as e:
                user_prompt = ""
            prompt = [
                {
                    "role": "system",
                    "content": system_prompt
                },
                {
                    "role" : "user",
                    "content" : user_prompt.format(type_of_error = type_of_error, error_explanation = info["error_explanation"], 
                                                   wrong_answer = info["wrong_answer"], question = info["question"], solution = info["answer"],
                                                   prev_injection_step = latest_wrong_step, injection_step = latest_wrong_step + 3)
                }
            ]

            #Get response from API call
            response = client.chat.completions.create(
                    model=model,
                    messages=prompt,
                    temperature=0.9
                )  
            message = response.choices[0].message
            if message is None or message.content is None:
                print(f"No valid message returned for question: {info["question"]}")
                continue    
            response_text = message.content.strip()
            response_text_extracted = None

            #Extract and convert to dictionary
            match = re.findall(r'\{.*\}', response_text, re.DOTALL)
            if match:
                response_text_extracted = match[0]
            else:
                print("NO JSON FORMAT OUTPUT FOUND! ERROR!")
                print(response_text)   
            try:
                response_dct = ast.literal_eval(response_text_extracted)
                wrong_solution_dct[question_idx] = response_dct
            except Exception as e:
                runtime_errors[question_idx] = response_text
                print("Found error at", question_idx, type_of_error, "error: ", e)
                print(response_text)
                sys.exit(0)
            #Save original response text for debugging
            wrong_solution_text[question_idx][type_of_error] = response_text

            wrong_solution_dct[question_idx]["all_wrong_steps"] = wrong_solution[question_idx]["wrong_steps"].copy()
            new_wrong_step = list(wrong_solution_dct[question_idx]["wrong_steps"].keys())[0]
            wrong_solution_dct[question_idx]["all_wrong_steps"][new_wrong_step] = wrong_solution_dct[question_idx]["wrong_steps"][new_wrong_step]

            wrong_solution_dct[question_idx]["all_explanations"] = wrong_solution[question_idx]["error_explanation"].copy()
            new_wrong_step = list(wrong_solution_dct[question_idx]["error_explanation"].keys())[0]
            wrong_solution_dct[question_idx]["all_explanations"][new_wrong_step] = wrong_solution_dct[question_idx]["error_explanation"][new_wrong_step]

            wrong_solution_dct[question_idx].pop("wrong_steps")
            wrong_solution_dct[question_idx].pop("error_explanation")
            #Progress tracker
            print(f"\rQuestion {question_idx} ({question_idx_num}/{question_len}), error type {error_idx}/{len(error_list)} generated", end = '', flush = True)

    #TODO: file name should also contain number of error types exist in the wrong solution
    #Save to file
    with open(generated_solution_fpath, 'w') as file:
        json.dump(wrong_solution_dct, file, indent=4)

    break
```
