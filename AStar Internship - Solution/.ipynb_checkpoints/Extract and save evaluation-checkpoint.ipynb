```python
import json
import pandas

with open('./evaluation/wrong solution/dataset_2025-06-09_14_16_30/openai_gpt-4.1_evaluating.json', 'r') as file:
    my_dict = json.load(file)

my_reliable_dict = my_dict['openai/gpt-4.1']
```

```python
import pandas as pd
evaluated_error = []
question_list = []
ground_error = []

models = ["openai/gpt-4.1", "qwen/qwen3-14b", "anthropic/claude-3.7-sonnet","microsoft/phi-4"]

dct = {}

load_question_list=0
for model in models:
    load_question_list += 1
    model = model.replace("/","_")
    with open(f'./evaluation/wrong solution/dataset_2025-06-09_14_16_30/{model}_evaluating.json', 'r') as file:
        my_dict = json.load(file)
    my_dict = my_dict['openai/gpt-4.1']
  
    if load_question_list == 1:
        for key in my_dict.keys():
            for another_key in my_dict[key].keys():
                question_list.append(key)
                ground_error.append(another_key)
                if another_key not in list(my_dict[key].keys()):
                    evaluated_error.append("No response")
                else:
                    try:
                        evaluated_error.append(list(my_dict[key][another_key].get('type_of_error',{None : 'None'}).values())[0])
                    except Exception as e:
                        print(key, another_key, model)
                        print(list(my_dict[key][another_key].get('type_of_error',{None : 'None'}).values()))

        dct["Question"] = question_list
        dct["Error type"] = ground_error
        dct[model] = list(evaluated_error)
    else:
        evaluated_error = []
        for key in my_reliable_dict.keys():
            for another_key in my_reliable_dict[key].keys():
                if another_key not in list(my_dict[key].keys()):
                    evaluated_error.append("No response")
                else:
                    try:
                        evaluated_error.append(list(my_dict[key][another_key].get('type_of_error',{None : 'None'}).values())[0])
                    except Exception as e:
                        evaluated_error.append("Invalid evaluator response")
        dct[model] = list(evaluated_error)
```

```python
df = pd.DataFrame(dct)
df
```

```python
df.to_excel("generic error evaluations.xlsx", index=False)
```
